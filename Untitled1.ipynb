```python
import re
import random
import datetime
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from collections import defaultdict, Counter
%matplotlib inline
```

```python
# 3. Жасанды логтарды жасау
def generate_synthetic_logs():
    hosts = ['host1','host2','host3']
    users = ['alice','bob','carol','dave','root','svc_backup']
    ips = ['192.168.1.'+str(i) for i in range(2,60)]
    services = ['sshd','sudo','cron','nginx','app']
    start = datetime.datetime.now() - datetime.timedelta(hours=6)

    lines = []
    for i in range(2000):
        t = start + datetime.timedelta(seconds=random.randint(0,6*3600))
        host = random.choice(hosts)
        srv = random.choice(services)
        ip = random.choice(ips)
        user = random.choice(users)

        # Brute force simulation
        if random.random() < 0.03:
            for j in range(random.randint(5,12)):
                ts = t + datetime.timedelta(seconds=j*3)
                lines.append(f"{ts.strftime('%b %d %H:%M:%S')} {host} {srv}: Failed password for {user} from {ip} port {random.randint(1024,65000)}")
        else:
            typ = random.choices(['ok','failed','info'], [0.7,0.15,0.15])[0]
            if typ == 'failed':
                lines.append(f"{t.strftime('%b %d %H:%M:%S')} {host} {srv}: Failed password for {user} from {ip} port {random.randint(1024,65000)}")
            elif typ == 'ok':
                lines.append(f"{t.strftime('%b %d %H:%M:%S')} {host} {srv}: Accepted password for {user} from {ip} port {random.randint(1024,65000)}")
            else:
                lines.append(f"{t.strftime('%b %d %H:%M:%S')} {host} {srv}: Connection from {ip} to port {random.randint(20,9000)}")

    with open('synthetic_syslog.log','w') as f:
        f.write('\n'.join(lines))

    return len(lines)
```

```python
# 4. Логтарды парсинг жасау
def parse_logs():
    pattern_failed = re.compile(r'(?P<ts>\w{3}\s+\d+\s+\d{2}:\d{2}:\d{2})\s+(?P<host>\S+)\s+(?P<svc>[^:]+):\s+Failed password for (?P<user>\S+) from (?P<ip>\d+\.\d+\.\d+\.\d+) port (?P<port>\d+)')
    pattern_ok = re.compile(r'(?P<ts>\w{3}\s+\d+\s+\d{2}:\d{2}:\d{2})\s+(?P<host>\S+)\s+(?P<svc>[^:]+):\s+Accepted password for (?P<user>\S+) from (?P<ip>\d+\.\d+\.\d+\.\d+) port (?P<port>\d+)')
    pattern_conn = re.compile(r'(?P<ts>\w{3}\s+\d+\s+\d{2}:\d{2}:\d{2})\s+(?P<host>\S+)\s+(?P<svc>[^:]+):\s+Connection from (?P<ip>\d+\.\d+\.\d+\.\d+) to port (?P<port>\d+)')

    rows = []
    with open('synthetic_syslog.log','r') as f:
        for ln in f:
            ln = ln.strip()
            m = pattern_failed.match(ln)
            if m:
                d = m.groupdict(); d['type']='failed'; rows.append(d); continue
            m = pattern_ok.match(ln)
            if m:
                d = m.groupdict(); d['type']='success'; rows.append(d); continue
            m = pattern_conn.match(ln)
            if m:
                d = m.groupdict(); d['type']='conn'; rows.append(d); continue

    df = pd.DataFrame(rows)
    this_year = datetime.datetime.now().year
    df['timestamp'] = pd.to_datetime(df['ts'] + f' {this_year}', format='%b %d %H:%M:%S %Y')
    df = df.sort_values('timestamp').reset_index(drop=True)
    return df
```

```python
def detect_bruteforce(df, window_seconds=300, threshold=8):
    failed_df = df[df['type']=='failed'].copy()
    failed_df['ts_epoch'] = failed_df['timestamp'].astype(np.int64)//10**9

    alerts = []
    ips = failed_df['ip'].unique()

    for ip in ips:
        ip_times = failed_df[failed_df['ip']==ip]['ts_epoch'].values
        i=0; j=0
        n = len(ip_times)
        while i < n:
            while j < n and ip_times[j] - ip_times[i] <= window_seconds:
                j += 1
            count = j - i
            if count >= threshold:
                alerts.append({
                    'ip': ip,
                    'start': pd.to_datetime(ip_times[i], unit='s'),
                    'count': count,
                    'type':'bruteforce'
                })
                i = j
            else:
                i += 1

    return pd.DataFrame(alerts)
```

```python
def detect_portscan(df, window_seconds=120, threshold=10):
    conn_df = df[df['type']=='conn'].copy()
    conn_df['ts_epoch'] = conn_df['timestamp'].astype(np.int64)//10**9

    scan_alerts = []
    for ip in conn_df['ip'].unique():
        tmp = conn_df[conn_df['ip']==ip].sort_values('ts_epoch')
        ports = tmp['port'].astype(int).values
        times = tmp['ts_epoch'].values

        i=0; j=0
        n=len(times)
        while i<n:
            seen = set()
            j=i
            while j<n and times[j] - times[i] <= window_seconds:
                seen.add(int(ports[j]))
                j+=1
            if len(seen) >= threshold:
                scan_alerts.append({
                    'ip':ip,
                    'start':pd.to_datetime(times[i],unit='s'),
                    'distinct_ports':len(seen),
                    'type':'portscan'
                })
                i = j
            else:
                i += 1

    return pd.DataFrame(scan_alerts)
```

```python
# 7. Визуализация
def visualize_results(df):
    failed_df = df[df['type']=='failed'].copy()

    # Top IPs by failed attempts
    plt.figure(figsize=(10, 6))
    top_failed = failed_df['ip'].value_counts().nlargest(10)
    plt.bar(top_failed.index, top_failed.values)
    plt.title('Топ 10 IP мекенжайлары бойынша қате кіру әрекеттері')
    plt.xlabel('IP мекенжайы')
    plt.ylabel('Қате әрекеттер саны')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

    # Time series
    plt.figure(figsize=(12, 6))
    failed_ts = failed_df.set_index('timestamp').resample('5Min').size()
    plt.plot(failed_ts.index, failed_ts.values)
    plt.title('Уақыт бойынша қате кіру әрекеттері (5 минуттық интервал)')
    plt.xlabel('Уақыт')
    plt.ylabel('Қате әрекеттер саны')
    plt.tight_layout()
    plt.show()
```

```python
# 8. Негізгі функция
def main():
    print("=== Зертханалық жұмыс: Лог мониторинг ===")

    # Логтарды жасау
    print("1. Жасанды логтар жасалуда...")
    log_count = generate_synthetic_logs()
    print(f"   {log_count} жолдық лог файлы жасалды")

    # Парсинг
    print("2. Логтар парсинг жасалуда...")
    df = parse_logs()
    print(f"   {len(df)} оқиға табылды")

    # Детекция
    print("3. Brute-force шабуылдарын анықтау...")
    brute_alerts = detect_bruteforce(df)
    print(f"   {len(brute_alerts)} brute-force оқиғасы табылды")

    print("4. Port scan шабуылдарын анықтау...")
    scan_alerts = detect_portscan(df)
    print(f"   {len(scan_alerts)} port scan оқиғасы табылды")

    # Барлық дабылдарды біріктіру
    all_alerts = pd.concat([brute_alerts, scan_alerts], ignore_index=True)

    # Визуализация
    print("5. Нәтижелерді визуализациялау...")
    visualize_results(df)

    # Нәтижелерді сақтау
    print("6. Нәтижелерді сақтау...")
    all_alerts.to_csv('detected_alerts.csv', index=False)
    df.to_csv('parsed_logs.csv', index=False)

    print("7. Нәтижелер:")
    print(f"   - Барлық оқиғалар: {len(df)}")
    print(f"   - Қате кіру әрекеттері: {len(df[df['type']=='failed'])}")
    print(f"   - Сәтті кіру әрекеттері: {len(df[df['type']=='success'])}")
    print(f"   - Қосылым оқиғалары: {len(df[df['type']=='conn'])}")
    print(f"   - Анықталған дабылдар: {len(all_alerts)}")

    if len(all_alerts) > 0:
        print("\nАнықталған дабылдар:")
        print(all_alerts.to_string(index=False))

# Бағдарламаны іске қосу
if __name__ == "__main__":
    main()
```
